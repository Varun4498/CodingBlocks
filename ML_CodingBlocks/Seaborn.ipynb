```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.barplot(x = 'sex',y = 'total_bill', data = tips ,estimator=np.std)
```

```python
sns.countplot(x = 'sex' , data=tips)
```

```python
sns.countplot(x = 'time' , data=tips)
```

```python
sns.boxplot(y='total_bill',x='day',data=tips)
```

```python
sns.boxenplot(y='total_bill',x='day',data=tips,hue='smoker')
```

```python
sns.violinplot(y='total_bill',x='day',data=tips,split=True)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)# kde - Kernel Density Estimation
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips , kind='kde') #kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex'
```

```python
sns.pairplot(tips,hue='sex')
```

```python
data = df.values #gives corresponding numpy array of the dataset
np.random.shuffle(data)
```

```python
print(data)
```

```python
X = data[ : ,1: ]
Y = data[ : ,0 ]
```

```python
print(X.shape,Y.shape)
```

```python
# Visualisation
```

```python
def drawImg(X,Y,i):
    plt.imshow(X[i].reshape(28,28),cmap='gray')
    plt.title("Label" + str(Y[i]))
    plt.show()
for i in range(1):
    drawImg(X,Y,i)
```

```python
## Split the dataset for training & testing
split = int(0.80*X.shape[0])
print(split)
```

```python
X_train,Y_train = X[ :split, : ],Y[:split]
X_test,Y_test = X[split:, : ],Y[split: ] #Starting from the split till the last
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
# Randomiziation of the array
import numpy as np
a = np.array([1,2,3,4,5])
np.random.shuffle(a)
print(a)
```

```python
# Try to plot a visualisation (Grid of first 25 images 5X5 Grid)
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(X_train[i].reshape(28,28),cmap='gray')
    plt.title(Y_train[i])
    plt.axis("off")
```

```python
# We can use sklearn lib for train test & split
from sklearn.sklearn.model_selection import train_test_split
XT,Xt,YT,Yt = train_test_split(X,Y,test_size = 0.2)#20 percent
print(XT.shape,YT.shape)
print(Xt.shape,Yt.shape)
```

```python
titles = list(df.get('movie_title'))
print(titles)
```

```python
scores = list(df.get('imdb_score'))
print(scores)
```

```python
for t in titles:
    print(t)
```

```python
freq_title = {}
for t in titles:
    length = len(t)
    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1
print(freq_title)
```

```python
X = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel('Movie Title Len')
plt.ylabel('No. Of Movies')
plt.show()
```

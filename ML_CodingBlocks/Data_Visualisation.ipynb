```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/varun/Desktop/ML & AI/ML_CodingBlocks/movie_metadata.csv")
```

```python
df.head(n=5)
```

```python
titles = df.get("movie_title")
print(titles)
```

```python
freq_titles = {

}
for title in titles:
    length = len(title)

    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie title")
plt.ylabel("Freq Count")
plt.title("Movie Data Visualisation")
plt.show()
```

```python
#Bivariant Variable
mean = np.array([0.0,0.0])
cov = np.array([[1,0.5],[0.5,1]])
dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
#Bivariant Variable
mean = np.array([0.0,0.0])
cov = np.array([[1,0.5],[0.5,1]])

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1,0],[0,3]])
dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])

plt.show()
```

```python
#Normalization Of Data
#Mean should be Zero
#Standard Deviation should be One
```

```python
dfx = pd.read_csv("/Users/varun/Desktop/ML & AI/ML_CodingBlocks/linearX.csv")
dfy = pd.read_csv("/Users/varun/Desktop/ML & AI/ML_CodingBlocks/linearY.csv")

x = dfx.values
y = dfy.values
#Normalize
x2 = (x-x.mean())/x.std()
plt.scatter(x,y,label = "Raw Data")
plt.scatter(x2,y,color = 'purple',label = "Normalized Data")
plt.legend()
plt.show()
```
